#### Exercise 2 : Create a plot with 4 sub-panels, use above code as a template

```python
import matplotlib.pyplot as plt
import numpy as np

# Create data
x = [1, 2, 3, 4]
y1 = [0, 0.5, 1, 0.2]
y2 = np.random.random(4) * 10   ## Gets 4 rand nums, scale by 10
y3 = np.random.random(4) * 10
y4 = np.random.random(4) * 10
nrow = 2
ncol = 2
y = [y1,y2,y3,y4]
color = ['red','blue','black','orange']


# Plot
fig = plt.figure()
gs = fig.add_gridspec(nrow, ncol)    # args = nrow,ncol
#gs.update(wspace=0.025, hspace=0.05)   # Adjust space between plots
yidx=0

for i in range(nrow):
    for j in range(ncol):
        ax = fig.add_subplot(gs[i,j])
        # Left Plot
        ax.plot(x, y[yidx], label="bad {} data".format(color[yidx]), color=color[yidx])
        ax.set_xlabel("My cool variable")
        ax.set_ylabel("My {} response".format(color[yidx]))
        ax.legend(loc='lower right', fontsize=10)
        ax.set_title("{}x worse data".format(yidx))
        yidx = yidx + 1

# Can adjust white space with below
fig.subplots_adjust(wspace=0.5,hspace=1)
#fig.tight_layout()
```

#### Exercise 3 : Create a with a colorbar, use above code as a template.  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Import data -
#   --> CC BY-SA 4.0 : https://archive.ics.uci.edu/dataset/109/wine
#   --> Aeberhard,Stefan and Forina,M.. (1991). Wine. UCI Machine Learning Repository. https://doi.org/10.24432/C5PC7J.
df = pd.read_csv("data/wine.data")       # No headers...
# Got headers from data/wine.names
header=["cult", "alcohol", "malic", "ash", "alcalinity", "mg", "phenol", "flav",
        "nonflav","proanth", "color", "hue", "od280", "proline"]
df.columns = header
cult1 = df[df['cult'] == 1].iloc[0:4]  # *.iloc for indices, *.loc for row names
cult2 = df[df['cult'] == 2].iloc[0:4]
cult3 = df[df['cult'] == 3].iloc[0:4]
allcult = [cult1, cult2, cult3]
cultname=['1','2','3']

# Create matrix for plotting
m = np.zeros([3, cult1.shape[0]])  # 3 cultivars, 5 wines per cultivar
# cultivars first
for i in range(len(allcult)):
    # wines next
    for j in range(allcult[i].shape[0]):
        m[i,j] = allcult[i]['alcohol'].iloc[j]

# Plot
fig, ax = plt.subplots()
im = ax.imshow(m)
fig.colorbar(im, ax=ax, anchor=(0, 0.3), shrink=0.7)

# y-axis
ax.set_yticks(np.arange(len(cultname)))
ax.set_yticklabels(cultname)
ax.set_ylabel("Cultivar")
# x-axis
ax.set_xticks(np.arange(m.shape[1]))
#ax.set_xticklabels(cultname)
ax.set_xlabel("Wine")
ax.legend(loc='lower right', fontsize=10)
ax.set_title("Alcohol content of 3 cultivars")
plt.show()      # Needed if plotting python outside of Jupyter notebook
```
